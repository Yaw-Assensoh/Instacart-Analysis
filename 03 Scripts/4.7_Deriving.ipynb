## Update the "busiest_day" column to "busiest _days



```python
# Count orders by day of the week
orders_by_day = df_orders_products_combined['order_dow'].value_counts ()
```

```python
# Identify the two busiest days
busiest_days = orders_by_day.nlargest(2).index.tolist()
```

```python
# Create the new 'busiest_days' column
df_orders_products_combined['busiest_days'] = df_orders_products_combined['order_dow'].apply(lambda x: 'Busiest days' if x in busiest_days else 'Other') 
```

## Create a new column for the two slowest days

```python
# Identify the two slowest days
slowest_days = orders_by_day.nsmallest(2).index.tolist()
```

```python
print("Busiest Days Column:/n", df_orders_products_combined['busiest_days'].value_counts())
```

```python
# create a new column for slowest days
df_orders_products_combined['slowest_days'] = df_orders_products_combined['order_dow'].apply(lambda x: 'Slowest_days' if x in slowest_days else 'Other') 
```

```python
print("/nSlowest Days Column:/n",df_orders_products_combined['slowest_days'].value_counts())
```

## Create "busiest_period_of_day" column

```python
# Count orders by hour
orders_by_hour = df_orders_products_combined['order_hour_of_day'].value_counts()
```

```python
# Create thresholds for Most, Average and Fewest orders
high_threshold = orders_by_hour.quantile(0.66)

low_threshold = orders_by_hour.quantile(0.033)
```

```python
# Map each hour to a category
def label_hour(hour):
    count = orders_by_hour[hour]
    if count >= high_threshold:
        return 'Most orders'
    elif count <= low_threshold:
        return 'Fewest orders'
    else:
        return 'Average orders'

df_orders_products_combined['busiest_period_of_day'] = df_orders_products_combined['order_hour_of_day'].apply(label_hour)
```

```python
print("Busiest Period of Day column:/n", df_orders_products_combined['busiest_period_of_day'].value_counts())
```

## Finding busiest hours of the day

```python
# busiest hours of the day
hourly_orders = df_orders_products_combined['order_hour_of_day'].value_counts().sort_index()
```

```python
# Finding busiest hour
busiest_hour = hourly_orders.idxmax()

busiest_order_count = hourly_orders.max()

print(f"the busiest hour of the day is{busiest_hour}:00 with {busiest_order_count} orders.")
```

## Observations
* The two busiest days based on oder volum are Sunday and Monday.
* The two slowest days are Wednesday and Thursday.
